## map et filter

```python
seq = [1,2,3,4,5]
```

```python
list(map(lambda num: num**2,seq))
```

```python
list(map(times2,seq))
```

```python
list(map(lambda var: var*2,seq))
```

```python
filter(lambda item: item%2 == 0,seq)
```

```python
list(filter(lambda item: item%2 == 0,seq))
```
